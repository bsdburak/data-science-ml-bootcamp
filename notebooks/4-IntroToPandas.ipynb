```python
# If two Series have the same indexes, basic operations such as sum() can be applied.
```

```python
series1 = pd.Series([10, 20, 30, 40], ["a", "b", "c", "d"])
```

```python
series1
```

```python
series2 = pd.Series([1, 2, 3, 4], ["a", "c", "e", "f"])
```

```python
series2
```

```python
series1 + series2
```

```python
series1 - series2
```
