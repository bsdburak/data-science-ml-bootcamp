```python
# multi index
```

```python
first_index = ["Simpson", "Simpson", "Simpson", "South Park", "South Park", "South Park"]
```

```python
inner_index = ["Homer", "Bart", "Marge", "Cartman", "Kenny", "Kyle"]
```

```python
zipped_index = list(zip(first_index, inner_index))
```

```python
zipped_index
```

```python
zipped_index = pd.MultiIndex.from_tuples(zipped_index)
```

```python
zipped_index
```

```python
samples_values = np.ones((6, 2))
```

```python
new_frame = pd.DataFrame(samples_values, index=zipped_index, columns=("A","B"))
```

```python
new_frame
```

```python
new_frame.loc["Simpson"]
```

```python
new_frame["A"]
```
