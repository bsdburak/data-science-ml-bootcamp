```python
def salary_category(salary):
    if salary < 50000:
        return "Low"
    elif 50000 <= salary < 80000:
        return "Medium"
    else:
        return "High"
```

```python
df["Salary"].apply(salary_category)
```

```python
df["Salary_Category"] = df["Salary"].apply(salary_category)
```

```python
df
```

```python
# if experience > 10 -> performance_score += 1 
```

```python
def adjust_performance(row):
    if row["Experience"] > 10:
        return row["Performance_Score"] + 1
    else:
        return row["Performance_Score"]
```

```python
df["Adjusted_Score"] = df.apply(adjust_performance, axis=1)
```
